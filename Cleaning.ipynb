```python
from dataprep.clean import clean_country
```

```python
import pandas as pd
```

```python
df = pd.read_excel("/content/Case_Study_to_clean.xlsx",index_col=0)
```

```python
df['geonameen'] = df['Country_clean']
```

```python
df2 = df.drop(['Country','Country_clean','Address'],axis=1)
```

```python
excel_table = pd.read_excel("/content/CaseStudy_Sourcing_sample.xlsx")
```

```python
concatenated_df = pd.concat([excel_table, df2])
concatenated_df
```
